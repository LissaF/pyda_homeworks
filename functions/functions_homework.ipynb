````python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
````

````python
def get_key(dictionary, value):
    for key, values in dictionary.items():
        if value in values:
            return key
````

````python
def owner_number():
    d_number = input('Введите номер документа: \n')
    for document in documents:
        if document['number'] == d_number:
            return print('Владелец документа:'+ document['name'])
    return print('Документ не найден в базе')    
````

````python
def directory_number():
    document_number = input('Введите номер документа: \n')
    if get_key(directories, document_number):
         return print('Документ хранится на полке: ' +  get_key(directories, document_number))        
    return print('Документ не найден в базе')    
````

````python
def all_information():
    for document in documents:
        print(f" №:{document['number']}, тип: {document['type']}, владелец: {document['name']}, полка хранения: {get_key(directories, document['number'])}")       
````

````python
def add_shelf():    
    shelf_number = input('Введите номер полки: \n')
    if shelf_number in directories:
        return print(f"Такая полка уже есть. Текущий перечень полок: {print_shelf_numbers()}")
    else:
        directories[shelf_number] = []
        return print(f"Полка добавлена. Текущий перечень полок: {print_shelf_numbers()}")  
````

````python
def print_shelf_numbers():
  return (', '.join(directories.keys()))
````

````python
def del_shelf():
  shelf_number = input('Введите номер полки: \n')
  if shelf_number not in directories.keys():
    return print(f"Такой полки не существует. Текущий перечень полок: {print_shelf_numbers()}")
  elif directories[shelf_number]:
    return print(f"На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {print_shelf_numbers()}")
  else:
    del(directories[shelf_number])
    return print(f"Полка удалена. Текущий перечень полок: {print_shelf_numbers()}")
````

````python
def add_document():
  d_number = input('Введите номер документа: \n')
  document_type = input('Введите тип документа: \n')
  document_owner = input('Введите владельца документа: \n')
  shelf_number = input('Введите номер полки: \n')
  if shelf_number not in directories.keys():
    print(f"Такой полки не существует. Добавьте полку командой ads.")
    print(f"Текущий список документов:")
    return all_information()
  else:
    documents.append({'type': document_type, 'number': d_number, 'name': document_owner}) 
    directories[shelf_number].append(d_number)
    print(f"Документ добавлен. Текущий список документов:")
    return all_information()
````

````python
def del_document():
  d_number = input('Введите номер документа: \n')
  for i in range(len(documents)):
    if documents[i]['number'] == d_number:
      del(documents[i])
      print('Документ удален. Текущий список документов:')
      return all_information()
  print('Документ не найден в базе')
  print('Текущий список документов:')
  return all_information()
    
````

````python
def change_shelf():
    d_number = input('Введите номер документа: \n')
    shelf_number = input('Введите номер полки: \n')
    if shelf_number not in directories.keys():
      return print(f"Такой полки не существует. Текущий перечень полок: {print_shelf_numbers()}")
    elif get_key(directories, d_number):
      directories[get_key(directories, d_number)].remove(d_number)
      directories[shelf_number].append(d_number)
      print('Документ перемещен.')
      print('Текущий список документов:')
      return all_information()
    else:
      print('Документ не найден в базе')
      print('Текущий список документов:')
      return all_information()
````

````python
def main():
  while True:
      cmd = input('Введите команду:\n')
      if cmd == 'q':
          break
      elif cmd == 'p': 
          owner_number()
      elif cmd == 's':
          directory_number()
      elif cmd == 'l':
          all_information()
      elif cmd == 'ads':
          add_shelf()
      elif cmd == 'ds':
          del_shelf()
      elif cmd == 'ad':
        add_document()
      elif cmd == 'd':
        del_document()
      elif cmd == 'm':
        change_shelf()
      else: raise ValueError
````

````python
main()
````
